```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import cv2
from PIL import Image
from random import randrange
from PIL import Image, ImageFont, ImageDraw, ImageEnhance
from random import randint
import pandas as pd
```

```python
def prepare_documents(stamps_dir, documents_dir):
    stamps = glob.glob(STAMPS_DIR+'*.png')
    documents = glob.glob(DOCUMENTS_DIR+'*.png')
    counter = 1
    columns = ['xmin','ymin','xmax','ymax']
    data = []
    for i in range(250):
            indexS = randint(0,len(stamps)-1)
            indexD = randint(0,len(documents)-1)
            background = Image.new('RGBA', (1632,2302),(255,255,255,0))
            document = Image.open(documents[indexD], 'r')
            stamp = Image.open(stamps[indexS], 'r')
            stamp = stamp.convert("RGBA")
            xmin,ymin,xmax,ymax = put_image(counter, stamp, document)
            data.append([xmin,ymin,xmax,ymax])
            counter = counter + 1
    df = pd.DataFrame(data=data, columns = columns)
    df.to_csv('coordinates.csv')
            
def rand_pos(bg_w, bg_h):
    x = randrange(bg_w)
    diff_x = bg_w - x
    y = randrange(bg_h)
    diff_y = bg_h - y
    return diff_x, diff_y, x, y

def put_image(counter, stamp, document):
    background = Image.new('RGBA', (1632,2302),(255,255,255,0))
    background2 = Image.new('RGBA', (1632,2302),(255,255,255,0))
    bg_w, bg_h = background.size
    diff_x, diff_y, x, y = rand_pos(bg_w, bg_h)
    if(stamp.width > background.width/2 or stamp.height > background.height/2):
        num = randint(1,3)
        stamp = stamp.resize((stamp.width//num, stamp.height//num), Image.ANTIALIAS)
    st_w, st_h = stamp.size
    shape = [(x,y), (x+st_w, y+st_h)]
    if(diff_x > st_w and diff_y > st_h):
        background.paste(stamp, (x,y), stamp)
        background = background.convert("RGB")
        background.save("../gold_data/gold_"+str(counter)+".png", "PNG")
        document.paste(stamp, (x,y), stamp)
        document.save("../data/doc_"+str(counter)+".png", "PNG")
        draw = ImageDraw.Draw(background2)
        r = draw.rectangle(shape, fill="black")
        background2 = background2.convert("RGB")
        background2.save("../map_data/doc_"+str(counter)+".png", "PNG")
    else:
        put_image(counter, stamp, document)
    return(x,y,x+st_w, y+st_h)
```

```python
def make_transparent(img_location):
    img = Image.open(img_location)
    img = img.convert("RGBA")
    datas = img.getdata()

    newData = []
    for item in datas:
        if item[0] == 255 and item[1] == 255 and item[2] == 255:
            newData.append((255, 255, 255, 0))
        else:
            if item[0] > 150:
                newData.append((0, 0, 0, 0))
            else:
                newData.append(item)


    img.putdata(newData)
    img.save("stamp_transparent.png", "PNG")
```
